```python
import uuid   # Unique identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join(r'C:\Users\Harsh Modi\yolov5\data', 'images') #/data/images
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)

    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))

        # Webcam feed
        ret, frame = cap.read()

        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')

        # Writes out image to file
        cv2.imwrite(imgname, frame)

        # Render to the screen
        cv2.imshow('Image Collection', frame)

        # 2 second delay between captures
        time.sleep(2)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import cv2
import numpy as np
import time
import threading
import pygame  # Use pygame for audio control


pygame.mixer.init()


warning_sound = pygame.mixer.Sound(r'C:\Users\Harsh Modi\Downloads\warning.mp3')

# Function to play alarm sound in a loop
def play_alarm():
    while not stop_alarm_flag:
        if not pygame.mixer.get_busy():
            warning_sound.play()


model = torch.hub.load('ultralytics/yolov5', 'custom', path=r'C:\Users\Harsh Modi\Downloads\last.pt', force_reload=True)


drowsy_start_time = None
alarm_thread = None
alarm_triggered = False
stop_alarm_flag = False
drowsy_threshold = 4

# Start the video capture
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()


    results = model(frame)

    labels = results.names
    detected_classes = results.pred[0][:, -1].cpu().numpy()


    detected_labels = [labels[int(cls)] for cls in detected_classes]
    print(f"Detected classes: {detected_labels}")

    # Check if drowsy is detected
    if 'drowsy' in detected_labels:
        if drowsy_start_time is None:
            drowsy_start_time = time.time()
            print("Drowsy detected, starting timer...")
        elif time.time() - drowsy_start_time > drowsy_threshold and not alarm_triggered:
            print("Drowsy for 4 seconds, playing alarm...")
            stop_alarm_flag = False  # Reset flag to allow alarm to play
            alarm_thread = threading.Thread(target=play_alarm)  # Play the alarm sound in a separate thread
            alarm_thread.start()
            alarm_triggered = True  # Ensure alarm only triggers once
    else:
        # Awake detected, stop alarm and reset drowsy timer
        if alarm_triggered:
            print("Awake detected, stopping alarm...")
            stop_alarm_flag = True  # Stop the alarm
            warning_sound.stop()  # Stop the sound immediately
            alarm_triggered = False  # Reset alarm trigger for next drowsy state
        drowsy_start_time = None  # Reset drowsy timer


    cv2.imshow('YOLO', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()

# Join the alarm thread if it's still running
if alarm_thread is not None:
    stop_alarm_flag = True  # Ensure alarm stops when exiting
    alarm_thread.join()
```
